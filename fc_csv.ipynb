```python
import pandas as pd
from tkinter import Tk
from tkinter.filedialog import askopenfilename, asksaveasfilename

# Open a file explorer to find the file
Tk().withdraw()  # we don't want a full GUI, so keep the root window from appearing
file_path = askopenfilename(title="Select the Excel file", filetypes=[("Excel files", "*.xlsx *.xls")])

# Read the Excel file
df = pd.read_excel(file_path, sheet_name='Base Forecast', header=1, usecols=['Último Eslabón', 'Canal','Marca','Segmentación Inchcape',"Forecast Colaborado" ,"Forecast Base'", 'Input', 'Forecast Colaborado','f2', 'f3','f4','f5','f6','f7','f8','f9','f10','f11','f12'])
# Use the first row as header
df.columns = df.iloc[0]
df = df[1:]

df.rename(columns={df.columns[3]:'Forecast Base', df.columns[6]:'Forecast Colaborado'},inplace=True)
# Open a file explorer to save the dataframe as a CSV
save_path = asksaveasfilename(defaultextension=".csv", filetypes=[("CSV files", "*.csv")], title="Save the dataframe as CSV")
df.to_csv(save_path, index=False)
columnas_din_fc = ['Último Eslabón','Marca','Segmentación Inchcape'] + [col for col in df.columns][6:]
df_din = df[columnas_din_fc]
import datetime 

meses = {
    '01': 'ene', '02': 'feb', '03': 'mar', '04': 'abr',
    '05': 'may', '06': 'jun', '07': 'jul', '08': 'ago',
    '09': 'sep', '10': 'oct', '11': 'nov', '12': 'dic'
}

# Obtener el mes actual
mes_actual = datetime.datetime.today().month 

# Crear nueva lista de nombres de columnas
nuevos_nombres = list(df_din.columns[:3])  # Mantener las 3 primeras columnas sin cambios

# Generar nombres de columna con meses
for i, _ in enumerate(df_din.columns[3:]):  
    mes_num = ((mes_actual + i - 1) % 12) + 1  # Calcular mes cíclicamente
    nombre_mes = f'Fc - {meses[str(mes_num).zfill(2)]}'
    nuevos_nombres.append(nombre_mes)

# Asignar nuevos nombres a df_din
df_din.columns = nuevos_nombres




```
